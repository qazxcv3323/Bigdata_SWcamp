```python
def searchdata(directory, output=[]):
    filelist = os.listdir(directory)
#     print(filelist)

    return filelist
    
weather_list = searchdata(os.path.join(os.getcwd(), weather_path))
weather_list
```

```python
w_2019 = pd.read_csv(weather_path + weather_list[-4], encoding='cp949')
w_2019.columns
```

```python
allData = [] # 읽어 들인 csv파일 내용을 저장할 빈 리스트를 하나 만든다

for file in range(1, len(weather_list)):
    df = pd.read_csv(weather_path + weather_list[file], encoding='cp949')
    allData.append(df) # 빈 리스트에 읽어 들인 내용을 추가한다

df_air = pd.concat(allData, axis=0, ignore_index=True) # concat함수를 이용해서 리스트의 내용을 병합
# axis=0은 수직으로 병합함. axis=1은 수평. ignore_index=True는 인데스 값이 기존 순서를 무시하고 순서대로 정렬되도록 한다.

df_air = df_air[['지점', '지점명', '일시', '기온(°C)', '강수량(mm)', '풍속(m/s)', '풍향(16방위)', '습도(%)', '증기압(hPa)', '이슬점온도(°C)', '현지기압(hPa)',
       '해면기압(hPa)', '지면온도(°C)', '5cm 지중온도(°C)', '10cm 지중온도(°C)', '20cm 지중온도(°C)', '30cm 지중온도(°C)']] 
df_air # 78399 rows × 17 columns
```
