```python
# 매출액/총자산(=자산총계)

과목별과목금액결측치평균으로제거.columns
재무 = pd.DataFrame([])



# 총자산증가율	"(당기말총자산/전기말총자산)*100-100 = (당기말 자산총계/전기말 자산총계)*100-100
# 자산총계 =총자본 = 총자산 =부채총계+자본총계(자기자본)

# 과목별과목금액결측치평균으로제거['총자산회전율'] = 과목별과목금액결측치평균으로제거['매출액'] - 과목별과목금액결측치평균으로제거['자본총계']

# 당기매출액 : 기업이 일정 기간 동안 얻은 모든 수익에서 지출한 모든 비용을 공제하고 순수하게 이익으로 남은 몫
# 매출액증가율 : 전년에 비해 얼마나 매출액이 증가했는지를 보여주는 성장성 지표
# 매출액증가율	(당기매출액/전기매출액)*100-100
# 과목별과목금액결측치평균으로제거['매출액증가율'] = 

# 유동비율	(유동자산/유동부채)*100
과목별과목금액결측치평균으로제거['유동비율'] = (과목별과목금액결측치평균으로제거['유동자산'] / 과목별과목금액결측치평균으로제거['유동부채']) * 100 #

# 차입금의존도	((차입금+회사채+유동성장기부채)/자본총계)*100
# 과목별과목금액결측치평균으로제거['차입금의존도'] =  

# 자기자본비율	"(자기자본(=자본총계)/총자본(=자산총계))*100
과목별과목금액결측치평균으로제거['자기자본비율'] = (과목별과목금액결측치평균으로제거['자본총계'] / 과목별과목금액결측치평균으로제거['자산총계']) * 100

# 부채비율	((유동부채+비유동부채)/자본총계(=자기자본))*100
과목별과목금액결측치평균으로제거['부채비율'] = ((과목별과목금액결측치평균으로제거['유동부채'] + 과목별과목금액결측치평균으로제거['비유동부채']) / 과목별과목금액결측치평균으로제거['자본총계']) * 100

# 이자보상비율	(영업이익/이자비용)*100
과목별과목금액결측치평균으로제거['이자보상비율'] = (과목별과목금액결측치평균으로제거['영업이익'] / 과목별과목금액결측치평균으로제거['이자비용']) * 100

# 적립금비율	((자본잉여금+이익잉여금)/자본총계(=자기자본))*100
과목별과목금액결측치평균으로제거['적립금비율'] = ((과목별과목금액결측치평균으로제거['자본잉여금'] + 과목별과목금액결측치평균으로제거['이익잉여금']) / 과목별과목금액결측치평균으로제거['자본총계']) * 100

# 총자본순이익율	(당기순이익(손실)/총자본(=자산총계))*100
과목별과목금액결측치평균으로제거['총자본순이익율'] = 과목별과목금액결측치평균으로제거['당기순이익(손실)'] / 과목별과목금액결측치평균으로제거['자산총계']#

# 매출영업이익율	(영업이익/매출액)*100
과목별과목금액결측치평균으로제거['매출영업이익률'] = round((과목별과목금액결측치평균으로제거['영업이익'] / 과목별과목금액결측치평균으로제거['매출액']) * 100 , 2)

# 총자산회전율	매출액/총자산(=자산총계)
과목별과목금액결측치평균으로제거['총자산회전율'] = 과목별과목금액결측치평균으로제거['매출액'] / 과목별과목금액결측치평균으로제거['자산총계']#

# 매출채권회전율	매출액/매출채권(평균)
과목별과목금액결측치평균으로제거['매출채권회전율'] = 과목별과목금액결측치평균으로제거['매출액'] / 과목별과목금액결측치평균으로제거['매출채권'].mean()#

# 유동비율	(유동자산/유동부채)*100
과목별과목금액결측치평균으로제거['유동비율'] = (과목별과목금액결측치평균으로제거['유동자산'] / 과목별과목금액결측치평균으로제거['유동부채'])# 

# 당좌비율	(당좌자산/유동부채)*100
과목별과목금액결측치평균으로제거['당좌비율'] = (과목별과목금액결측치평균으로제거['당좌자산'] / 과목별과목금액결측치평균으로제거['유동부채']) * 100


재무 = 과목별과목금액결측치평균으로제거[['중분류항목명','조사년도','유동비율', '자기자본비율', '부채비율', '이자보상비율','적립금비율', '총자본순이익율','매출영업이익률', '총자산회전율','매출채권회전율', '유동비율', '당좌비율']]
재무
```
